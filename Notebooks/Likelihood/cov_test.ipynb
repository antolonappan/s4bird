```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pk
```

```python
cov_del_sim = pk.load(open('cov_del_r.pkl','rb'))
cov_del_sim_s4 = pk.load(open('cov_del_r_s4.pkl','rb'))
cov_del_idl = pk.load(open('cov_del.pkl','rb'))
cov_len = pk.load(open('cov_len.pkl','rb'))
```

```python
plt.figure(figsize=(8,8))
plt.loglog(np.sqrt(np.diag(cov_del_sim)),label='Realisation',ls=':',lw=3)
plt.loglog(np.sqrt(np.diag(cov_del_sim_s4)),label='Realisation_s4',ls='--',lw=3)
plt.loglog(np.sqrt(np.diag(cov_del_idl)),label='Ideal')
plt.loglog(np.sqrt(np.diag(cov_len)),label='Lensed')
plt.legend()
```
